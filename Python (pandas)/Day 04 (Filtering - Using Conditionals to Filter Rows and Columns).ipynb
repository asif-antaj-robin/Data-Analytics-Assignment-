# Filtering - Using Conditionals to Filter Rows and Columns

```python
import pandas as pd
```

```python
People = {
    "First Name" : ["Asif Antaj", "Afrin Jahan", "Afifa Jahan", "Asif Antaj"],
    "Last Name" : ["Robin", "Rupa", "Rafa", "Rayhan"],
    "Email" : ["asifantajrobin@gmal.com", "afrin@gmail.com", "afifa@gmail.com", "asif@gmail.com"]
}
```

```python
df = pd.DataFrame(People)
df
```

```python
df["First Name"] == "Asif Antaj"
```

```python
df[df["First Name"] == "Asif Antaj"]
```

```python
Filter = (df["First Name"] == "Asif Antaj")
Filter
```

```python
df[Filter]
```

```python
df.loc[Filter]
```

```python
df.loc[Filter, "Email"]
```

```python
df.loc[Filter, "Email"]
```

```python
Filter = (df["First Name"] == "Asif Antaj") & (df["Last Name"] == "Robin")
Filter
```

```python
df.loc[Filter]
```

```python
df.loc[Filter, "Email"]
```

```python
df.loc[~Filter, "Email"]
```

```python
df = pd.read_csv("employee_data.csv")
df.head()
```

```python
df.Salary
```

```python
High_Salary = (df["Salary"]>110000)
High_Salary
```

```python
df.loc[High_Salary]
```

```python
df.columns
```

```python
df.loc[High_Salary, ["Name", "City", "ProgrammingSkills", "Salary"]]
```

```python
df.City.value_counts()
```

```python
Citys = ["Dhaka", "Chittagong", "Rajshahi"]
filt = df["City"].isin(Citys)
filt
```

```python
df.loc[filt, "City"]
```

```python
df_filtered = df[High_Salary & filt]
df_filtered
```
