```python
country_respondents = df["Country"].value_counts()
country_respondents
```

```python
Country_uses_python = Country_grp["ProgrammingSkills"].apply(lambda x: x.str.contains("Python", case=False, na=False).sum())
Country_uses_python
```

```python
python_df = pd.concat([country_respondents, Country_uses_python], axis=1, sort=False)
python_df
```

```python
python_df.reset_index(inplace=True)
python_df
```

```python
python_df.rename(columns={
    "Country" : "Countrys",
    "ProgrammingSkills" : "Python_Uses"
}, inplace=True)

python_df
```

```python
python_df["PctKnowsPython"] = (python_df["Python_Uses"]/python_df["count"]) * 100
python_df
```

```python
python_df.sort_values(by = "PctKnowsPython", ascending = False, inplace=True)
python_df.head(50)
```
