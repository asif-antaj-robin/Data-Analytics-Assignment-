# Pie Charts

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(10, 5))

slices = [60, 40]
labels = ["Sixty", "Forty"]
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor' : 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(10, 5))

slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels = labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(10, 5))

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels = labels, shadow = True, wedgeprops = {'edgecolor' : 'black'}, explode = explode)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(10, 5))

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.2, 0]

plt.pie(slices, labels = labels, shadow = True, startangle=90, autopct='%1.1f%%', wedgeprops = {'edgecolor' : 'black'}, explode = explode)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.savefig("Pie_Charts.png")
plt.show()
```
